```python
# Для роботи з даними
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split

# Для роботи з Keras
from keras.models import Sequential # модель прямого розповсюдження
from keras.layers import Dense # повнозв'язні прошарки
```

```python
data = pd.read_csv('/content/music(1).csv')
data[:5]
```

```python
data = data.drop('filename', axis=1)
data[:5]
```

```python
data.info()
```

```python
data['label'].value_counts(normalize=True)
```

```python
data.isnull().sum()
```

```python
data['label'].unique()
```

```python
 def label_to_numeric(a):
     if a == 'blues':
         return 1
     if a == 'classical':
         return 2
     if a == 'country':
         return 3
     if a == 'disco':
         return 4
     if a == 'hiphop':
         return 5
     if a == 'jazz':
         return 6
     if a == 'metal':
         return 7
     if a == 'pop':
         return 8
     if a == 'reggae':
         return 9
     if a == 'rock':
         return 10

 data['Label'] = data['label'].apply(label_to_numeric)
 data.head()
```

```python
data['Label'].value_counts(normalize=True)
```

```python
X = data.drop(['Label','label'], axis = 1)
X
```

```python
Y = pd.get_dummies(data.Label, prefix='Label')
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.15)
```

```python
mean = X_train.mean(axis=0)
std = X_train.std(axis=0)

X_train -= mean
X_train /= std
X_test -= mean
X_test /= std
```

```python
model = Sequential()
```

```python
model.add(Dense(30, input_dim=X_train.shape[1], activation="relu"))
```

```python
model.add(Dense(20, activation="relu"))
```

```python
model.add(Dense(10, activation="softmax"))
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, Y_train, epochs=50, validation_split=0.15, batch_size=16,verbose=0)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Val'], loc='upper left')
plt.show()
```

```python
sample = X_test[9:10]
prediction = model.predict(sample)
print(prediction)
```

```python
score = np.max(prediction)
prediction = np.argmax(prediction)

print(f'Передбачено: {prediction} з достовірністю {score*100:.2f}%', f'Повинно бути: {np.argmax(Y_test[0:1])}', sep='\n')
```
